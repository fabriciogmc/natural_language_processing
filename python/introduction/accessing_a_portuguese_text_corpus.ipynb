```python
# In this example we load Macmorpho corpus
# to check if a word belongs to it:

# Example accessing text corpus in portuguese (Mac Morpho)
from nltk.corpus import mac_morpho as mm

# Let's check some workds and its tags:
print(mm.tagged_words()[:10])

# Now we check if some words belong to this corpora
my_words = ["táxi", "camiseta", "abbrre"]
for my_word in my_words:
    if my_word in mm.words():
        print(my_word + " belongs to MM")
    else:
        print(my_word + " does not belong to MM")
```

```python
# Now we get some statistics:
mm_words = mm.words()
print('Number of times the word taxi appears in Macmorpho:', mm_words.count('táxi'))
print('Number of times the word camiseta appears in Macmorpho: ', mm_words.count('camiseta'))
```
