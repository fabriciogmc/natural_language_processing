```python
import spacy
from spacy import displacy

sentence = "Donald Trump was the USA president in 2020"
print("Building dependency parse tree:")
nlp = spacy.load('en_core_web_sm')
my_document = nlp(sentence)
displacy.render(my_document, options={'distance':100, 'arrow_stroke': 1.5, 'arrow_width':8}, jupyter=True)

for token in my_document:
    print(token.text, "→ children:",
          [child.text for child in token.children])
```
