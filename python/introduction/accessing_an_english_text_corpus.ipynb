```python
# A simple example that access a english text corpus and
# checks some synonyms for the chosen word.
#
# Author: Fabrício Galende Marques de Carvalho

import nltk
# Uncomment the first time you execute this notebook
# nltk.download('wordnet')
# nltk.download('omw-1.4')

from nltk.corpus import wordnet

word = 'help'

# Getting synonyms:
word_synsets = wordnet.synsets(word)

for synset in word_synsets:
    print(("Synset Name: {name}\n"
           "POS Tag: {tag} \n"
           "Definition: {definition}\n"
           "Examples: {examples}\n").format(name=synset.name(),
                                           tag=synset.pos(),
                                           definition = synset.definition(),
                                           examples = synset.examples()))
```

```python
# Now let's get some lemmas (in this case, equivalent words, with respect to meaning) in other languages:

print("Language supported:", wordnet.langs(), "\n\n")

print(wordnet.synset('dog.n.01').lemma_names('ita'))
print(wordnet.synset('dog.n.01').lemma_names('por'))
```
